```python
def load_embeddings(path):
    """
    Load serialized embeddings from a binary object file.

    :param path: Path to the compressed archive file.
    :return: A list of lists, where each inner list contains embeddings for a batch within an epoch.
    """
    # Load the compressed archive
    compressed_archive = np.load(path, allow_pickle=True)

    # Extract the embeddings from the archive
    all_batch_embeddings = []
    for epoch_idx in sorted(compressed_archive.keys()):
        epoch_embeddings = compressed_archive[epoch_idx]
        all_batch_embeddings.append(epoch_embeddings)

    print(f"Embeddings loaded from compressed archive: {path}")
    # logger.info(f"Embeddings loaded from compressed archive: {path}")

    return all_batch_embeddings
```

```python
def load_embeddings(path):
    """
    Load serialized embeddings from a binary object file.

    :param path: Path to the compressed archive file.
    :return: A list of lists, where each inner list contains embeddings for a batch within an epoch.
    """
    # Load the compressed archive
    compressed_archive = np.load(path, allow_pickle=True)

    # Extract the embeddings from the archive
    all_batch_embeddings = []
    for epoch_idx in sorted(compressed_archive.keys()):
        epoch_embeddings = compressed_archive[epoch_idx]
        all_batch_embeddings.append(epoch_embeddings)

    print(f"Embeddings loaded from compressed archive: {path}")
    # logger.info(f"Embeddings loaded from compressed archive: {path}")

    return all_batch_embeddings

npz_files = [
    'models/1e0f422a-9064-4379-a3bb-d49ef138e31c.npz',
    'models/39813543-2407-4fe5-830d-04297b6577f5.npz',
    # 'models/5401c48e-13d1-46c5-a8ef-1c5db36f428e.npz',
    'models/ce5ef1e3-65bb-4e37-8ada-4b6b4f10c75c.npz',
    'models/cf213e76-5247-4c59-94ee-38adccdd9c4e.npz',
    'models/f88688ca-1e8c-4f39-ae48-92c2bd9ed6f5.npz'
]

# Iterate through each file
client_embeddings = []
for file in npz_files:
    # Load embeddings using the provided function
    embeddings = load_embeddings(file)
    
    # Print the required information
    print(f"Filename: {file}")
    print(f"Length: {len(embeddings)}")
    client_embeddings.append(embeddings)
    # print(f"First element: {embeddings[0] if embeddings else 'No embeddings'}")

# Using client_embeddings, make a client_embeddings_dict with client id as key and embeddings as value
client_embeddings_dict = {client_id+1: embeddings for client_id, embeddings in enumerate(client_embeddings)}


def merge_client_data(client_data):
    """
    Merge embeddings for each batch within an epoch into a single ndarray.

    :param client_data: A list of lists, where each inner list contains embeddings for a batch within an epoch.
    :return: A list of ndarrays, where each ndarray contains all embeddings for a given epoch.
    """
    # Single client
    # client_data = load_embeddings(file)
    # client_data[0][0].shape -> (256, 7)
    # Merge all batch ndarrays into one ndarray for each epoch
    merged_client_data = [np.vstack(epoch_data) for epoch_data in client_data]
    # Now, when calling merged_client_data[0], it will return an ndarray with size of dataset_size x 7
    # merged_client_data[0].shape -> (23526, 7)

    return merged_client_data

def make_x_input_tensor(client_embeddings_dict):
    # first_client = merge_client_data(client_embeddings[0])
    # all_clients_data = [first_client[0]]  # Start with the first client's data

    # other_clients_data = []
    # for client_data in client_embeddings[1:]:
    #     merged_client_data = merge_client_data(client_data)
    #     other_clients_data.append(merged_client_data)

    all_clients_data = []
    
    for client_id in sorted(client_embeddings_dict.keys()):
        client_data = client_embeddings_dict[client_id]
        merged_client_data = merge_client_data(client_data)
        all_clients_data.append(merged_client_data[0])

    # all_clients_data.extend([client_data[0] for client_data in other_clients_data])

    # Stack all data vertically
    final_data_np = np.hstack(all_clients_data)

    # Convert to torch tensor
    final_data_tensor = torch.from_numpy(final_data_np)
    
    return final_data_tensor, final_data_np

# Example usage:
# final_data_tensor = make_x_input_tensor(npz_files)
x_input, x_input_np = make_x_input_tensor(client_embeddings_dict)
print(x_input.shape)
```
